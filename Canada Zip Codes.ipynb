## Wiki scraping data

```python
import pandas as pd
import numpy as np
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
print('html read!')
```

## Saving the table of the html to the data frame


```python
df =  df[0]
df.head()
```

## cleaning the table

```python
df =  df[df['Neighbourhood'] != "Not assigned"]
df.head()
```

## Merging duplicate post codes

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df.head()
```
