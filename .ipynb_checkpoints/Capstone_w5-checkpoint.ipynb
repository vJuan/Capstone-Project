### Fucntion to explore the venues

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
SanFV = getNearbyVenues(names=dfSan['Neighbourhood'],
                                   latitudes=dfSan['Latitude'],
                                   longitudes=dfSan['Longitude']
                                  )
```

```python
SeaV = getNearbyVenues(names=dfSea['Neighbourhood'],
                                   latitudes=dfSea['Latitude'],
                                   longitudes=dfSea['Longitude']
                                  )
```

```python
# one hot encoding
san_onehot = pd.get_dummies(SanFV[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
san_onehot['Neighborhood'] = SanFV['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [san_onehot.columns[-1]] + list(san_onehot.columns[:-1])
san_onehot = san_onehot[fixed_columns]

san_onehot.head()
```

```python
# one hot encoding
sea_onehot = pd.get_dummies(SeaV[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
sea_onehot['Neighborhood'] = SeaV['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [sea_onehot.columns[-1]] + list(sea_onehot.columns[:-1])
sea_onehot = sea_onehot[fixed_columns]

sea_onehot.head()
```

### Gruoping for clustering

```python
sang = san_onehot.groupby('Neighborhood').mean().reset_index()
sang.head()
```

```python
seag = sea_onehot.groupby('Neighborhood').mean().reset_index()
seag.head()
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 9

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = sang['Neighborhood']

for ind in np.arange(sang.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(sang.iloc[ind, :], num_top_venues)

sans = neighborhoods_venues_sorted
sans.head()
```

```python
num_top_venues = 9

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = seag['Neighborhood']

for ind in np.arange(seag.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(seag.iloc[ind, :], num_top_venues)

seas = neighborhoods_venues_sorted
seas.head()
```

### Clustering

```python
# set number of clusters
kclusters = 9

sangc= sang.drop('Neighborhood', 1)

# run k-means clustering
kmeansSan = KMeans(n_clusters=kclusters, random_state=0).fit(sangc)

# check cluster labels generated for each row in the dataframe
kmeansSan.labels_[0:10] 
```

```python
# set number of clusters
kclusters = 9

seagc= seag.drop('Neighborhood', 1)

# run k-means clustering
kmeansSea = KMeans(n_clusters=kclusters, random_state=0).fit(sangc)

# check cluster labels generated for each row in the dataframe
kmeansSea.labels_[0:10] 
```

```python
# add clustering labels
#neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

sanm = dfSan
sanm['Cluster Labels'] = kmeansSan.labels_


sanm = sanm.join(sans.set_index('Neighborhood'), on='Neighborhood')

sanm.head() # check the last columns!
```

```python
add = 'San Francisco, California USA'
geo = Nominatim()
loc = geo.geocode(add)
lat = loc.latitude
lon = loc.longitude
df = dfSan
```

```python
# create map
map_clusters = folium.Map(location=[lat, lon], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(sanm['Latitude'], sanm['Longitude'], sanm['Neighbourhood'],sanm['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
```
